**Create labels for a plot**

```python
x= np.array([80,85,90,95])
y= np.array([20,60,90,120])

plt.plot(x,y)
plt.title('Artificial Intelligence')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x= np.array([80,85,90,95])
y= np.array([20,60,90,120])

plt.plot(x,y)
font1={'family':'arial','color':'blue','size':'20'}
font2={'family':'arial','color':'green','size':'20'}
plt.title('Artificial Intelligence', fontdict={'family':'arial', 'color':'blue', 'size':'20'})
plt.xlabel('x-axis', fontdict=font1)
plt.ylabel('y-axis', fontdict=font2)
```
